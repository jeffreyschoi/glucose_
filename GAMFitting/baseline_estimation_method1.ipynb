```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from csaps import csaps
import math
```

```python
start_times = [1, 2, 3, 5, 7] #Starting time of each activity bout goes here
end_times = [] #Ending time of each activity bout goes here
df = pd.read_csv("data.csv")
```

```python
# We want S(x, xi) to be the weight given do data point xi for an estimate at x
def gamma(x):
  prior_act = []
  for p in start_times:
    if p < x:
      prior_act.append(p)
  return x - max(prior_act)

def S(x, xi):
  sigma2 = 1
  r = 1
  return math.exp(-(xi - x)**2 / 2*sigma2 + r * gamma(xi))

def kernel_smoother(X, y):
  def y_hat(x):
    weighted_sum = 0
    for i in range(len(X)):
      xi = X[i]
      weighted_sum += S(x, xi) * y[i]

    sum_of_weights = sum([S(x, xi) for xi in X])
    return weighted_sum / sum_of_weights
  return y_hat

fitted_function = kernel_smoother(df['X'], df['y'])

X_domain = np.linspace(df['X'].min(), df['X'].max(), 100)
y_domain = fitted_function(X_domain)
plt.plot(X_domain, y_domain)
```
