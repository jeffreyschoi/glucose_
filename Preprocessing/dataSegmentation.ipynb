```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
ind = '006'

new_folder_name = "segmented"
new_folder_path = os.path.join(f'/content/drive/MyDrive/ResearchProject/datafiles/{ind}', new_folder_name)
os.makedirs(new_folder_path, exist_ok=True)

df = pd.read_csv(f'/content/drive/MyDrive/ResearchProject/datafiles/{ind}/ACC_{ind}.csv')
df['datetime'] = pd.to_datetime(df['datetime'])
df['day'] = df['datetime'].dt.day

grouped = df.groupby('day')

for day, group in grouped:
  filename = f'/content/drive/MyDrive/ResearchProject/datafiles/{ind}/segmented/ACC_{ind}_day{day}.csv'
  group.to_csv(filename, index=False)
  print(f'Saved {filename}')
```

```python
ind_list = []
for i in range(1, 17):
  if i < 10:
    ind_list.append(f'00{i}')
  else:
    ind_list.append(f'0{i}')

for ind in ind_list:
  new_folder_name = "segmented"
  new_folder_path = os.path.join(f'/content/drive/MyDrive/ResearchProject/datafiles/{ind}', new_folder_name)
  os.makedirs(new_folder_path, exist_ok=True)

  df = pd.read_csv(f'/content/drive/MyDrive/ResearchProject/datafiles/{ind}/ACC_{ind}.csv')
  df['datetime'] = pd.to_datetime(df['datetime'])
  df['day'] = df['datetime'].dt.day

  grouped = df.groupby('day')

  for day, group in grouped:
    filename = f'/content/drive/MyDrive/ResearchProject/datafiles/{ind}/segmented/ACC_{ind}_day{day}.csv'
    group.drop(columns = ['day'])
    group.to_csv(filename, index=False)
    print(f'Saved {filename}')
```
